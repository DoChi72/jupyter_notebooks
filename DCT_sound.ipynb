```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class DCT:
    def __init__(self,N):
        self.N = N    # データ数．
        # 1次元，2次元離散コサイン変換の基底ベクトルをあらかじめ作っておく
        self.phi_1d = np.array([ self.phi(i) for i in range(self.N) ])
        
        # Nが大きいとメモリリークを起こすので注意
        # MNISTの28x28程度なら問題ない
        self.phi_2d = np.zeros((N,N,N,N))
        for i in range(N):
            for j in range(N):
                phi_i,phi_j = np.meshgrid(self.phi_1d[i],self.phi_1d[j])
                self.phi_2d[i,j] = phi_i*phi_j
        #print(self.phi_2d)
        
    def dct1(self,data):
        """ 1次元離散コサイン変換を行う """
        return self.phi_1d.dot(data)
    
    def idct1(self,c):
        """ 1次元離散コサイン逆変換を行う """
        return np.sum( self.phi_1d.T * c ,axis=1)
    
    def dct2(self,data):
        """ 2次元離散コサイン変換を行う """
        return np.sum(self.phi_2d.reshape(N*N,N*N)*data.reshape(N*N),axis=1).reshape(N,N)
    
    def idct2(self,c):
        """ 2次元離散コサイン逆変換を行う """
        return np.sum((c.reshape(N,N,1)*self.phi_2d.reshape(N,N,N*N)).reshape(N*N,N*N),axis=0).reshape(N,N)
    
    def phi(self,k):
        """ 離散コサイン変換(DCT)の基底関数 """
        # DCT-II
        if k == 0:
            return np.ones(self.N)/np.sqrt(self.N)
        else:
            return np.sqrt(2.0/self.N)*np.cos((k*np.pi/(2*self.N))*(np.arange(self.N)*2+1))
        # DCT-IV(試しに実装してみた)
        #return np.sqrt(2.0/N)*np.cos((np.pi*(k+0.5)/self.N)*(np.arange(self.N)+0.5))
```

# 画像の読み込み

```python
from PIL import Image
```

```python
# 任意の座標を中心に、任意の正方形でくり抜く
def crop_anycoord(pil_img, coord, crop_width, crop_height):
    img_width, img_height = pil_img.size
    return pil_img.crop((coord[0] - (crop_width//2),
                                   coord[1] - (crop_height//2),
                                   coord[0] + (crop_width//2),
                                   coord[1] + (crop_height//2)))
```

```python
# 任意のサイズで画像を左上から正方形で刻む
def image_chop(img, cut_size):
    img_width, img_height = img.size
    chopped = []
    coord = [cut_size//2, cut_size//2]
    
    for j in range(img_height//cut_size):
        coord[0] = cut_size//2
        for i in range(img_width//cut_size):
            chopped.append(crop_anycoord(img, coord, cut_size, cut_size))
            coord[0] = coord[0] + cut_size
        coord[1] = coord[1] + cut_size
        
    return chopped
```

## 上記の工程を細切れにした画像のDCT変数の分繰り返す

```python
chopped_c_tones = []
for k in tqdm(range(len(gray_chopped_c))):
    #0番目がDC成分
    low_c = []
    cnt = 7
    num = 0
    for i in range(7):
        for j in range(cnt):
            low_c.append(gray_chopped_c[k][j, num])
        cnt =  cnt - 1
        num = num + 1
    
    high_c = []
    cnt = 1
    num = 1
    for i in range(7):
        for j in range(cnt):
            high_c.append(gray_chopped_c[k][7-j, num])
        cnt =  cnt +1
        num = num + 1
        
    low_c.pop(0)
    high_c.pop(21)
    
    low_index = np.where(gray_chopped_c[k] == max(low_c))
    high_index = np.where(gray_chopped_c[k] == max(high_c))

    chopped_c_tones.append([int(min(low_index)), 7-int(min(high_index))])
```

# 音を割り振る

```python
#Cmajor
def Cmajor(chopped_c_tones, list_length, assignmentFreq):
    n = 0
    for x in range(list_length):
        m = 0
        for y in range(len(chopped_c_tones)): #和音の数だけ繰り返す
            if 0 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 261.63
            elif 1 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 293.66
            elif 2 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 329.63
            elif 3 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 349.23
            elif 4 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 392.00
            elif 5 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 440.00
            elif 6 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 493.88
            elif 7 == chopped_c_tones[n][m]:
                assignmentFreq[n][m] = 523.25
            m = m+1
        n = n+1
    print(assignmentFreq)
    return assignmentFreq
```

```python
def createCombinedWave (A, freqList, fs, length):
    """freqListの正弦波を合成した波を返す"""
    data = []
    amp = float(A) / len(freqList)
    # [-1.0, 1.0]の小数値が入った波を作成
    for n in np.arange(length * fs):  # nはサンプルインデックス
        s = 0.0
        for f in freqList:
            s += amp * np.sin(2 * np.pi * f * n / fs)
        # 振幅が大きい時はクリッピング
        if s > 1.0:  s = 1.0
        if s < -1.0: s = -1.0
        data.append(s)
    return data
```
