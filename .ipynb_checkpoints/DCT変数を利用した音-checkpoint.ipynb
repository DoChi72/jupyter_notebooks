# 概要

DCT変数を用いた方法を用いた、音楽作りの経過を報告する。今回の画像の取り扱い方はCP-1に該当する。  
音の出し方は以前に打ち合わせした、低周波と高周波の値の大きい代表２つから２音の和音を一つのブロックから作る方法である。

# 画像の取り扱い・音の変換について


<div style="text-align: center;"><b>縮小し、細切れにした画像</b></div>


<div style="text-align: center;"><b>細切れにした画像それぞれのDCT変数のヒートマップ</b></div>（画面輝度を高くすると良く見える）

上図のDCT変数のブロック１つから、低周波と高周波の代表の値を件の方法で261.63〜523.25Hz（C4~C5）に当てはめた。  
音ファイルとしては、ブロック１つ２音の和音で1秒間。合計６４秒のファイルになっている。
