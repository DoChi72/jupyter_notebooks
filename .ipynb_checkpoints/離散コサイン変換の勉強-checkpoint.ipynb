# 離散コサイン変換の概要

　離散コサイン変換は離散信号を周波数領域へ変換する手法（離散フーリエ変換など）の一つである。入力信号を xi　( i = 0,1,2,‥,n-1)とする場合に$\hat{x}は入力信号$、$\hat{c}$DCT係数を、$T$を変換行列とした以下のような変換のことである。

$$\hat{c} = T \cdot \hat{x}$$

　変換ベクトル$T$は入力信号が以下の式で与えられている。

$$
T_{ij} = \sqrt{\frac{2}{n}} k_i cos
\biggl[\frac{(i-1)(j-1/2)}{n}\pi\biggr]
$$

$$
k_i =
\begin{cases}
\frac{1}{\sqrt{2}}&(i = 0)\\
1&(i \neq 0)
\end{cases}      
$$

　離散コサイン変換の逆変換は用意で逆行列は転置行列で与えられている。

$$T^{-1} = {}^tT$$

# 2次元離散コサイン変換

　画像信号を取り扱うために、拡張された2次元離散コサイン変化の式が以下のように与えられている。

$$C = T \cdot G \cdot {}^tT$$

　逆変換は次のように与えられる。

$$G = {}^tT \cdot C \cdot T$$

　離散コサイン変換は画像信号を、以下に示すような基底画像（基底関数）の重ね合わせで表現する"重み"の係数（どの周波数の画像の比重が多いか）を算出もの考えることができる。
