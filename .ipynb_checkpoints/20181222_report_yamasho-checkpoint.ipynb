# 画像の切り抜き方

* 指定した矩形で切り抜き、再構成する。（この方法だけですでに抜き出し方が先生のCP-3まである）
* 矩形で切り出した画像を重ねる。（CP-4）
* 物体の形状を検出し、一番大きい幅のエッジ間の長さの矩形で切り抜く。（CP-5）
* ヒストグラムの画像・波形を正方形に納めて出力し利用する。(CP-6)
