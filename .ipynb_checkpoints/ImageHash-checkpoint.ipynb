* ImageHashをインストールし、画像をハッシュ値で表すサンプルコードを実行した
* 続けてImageHashライブラリ、pHashについての勉強を始める

```python
from PIL import Image
import imagehash
```

```python
hash = imagehash.whash(Image.open(r'data/lena.png'))
```
