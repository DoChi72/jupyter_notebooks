# 画像の類似度の算出のテストコード

```python
lenna = PIL.Image.open('./data/DWT_lenna.png')
lenna1 = PIL.Image.open('./data/DWT_lenna1.jpg')
```

```python
lenna
```

* オリジナルサイズ(lenna)

```python
lenna1
```

* 512x512>400x400のリサイズをし、カラースキーマがRGB>YMCKと変換され、JPEG圧縮をかけた画像(lenna1)

## pHash値を計算し比較する

```python
h = imagehash.phash(lenna)
h1 = imagehash.phash(lenna1)

print(h-h1)
```

```python
print(h) #オリジナルのハッシュ値
```

```python
print(h1)#変換後のハッシュ値
```

```python
(h - h1)/len(h.hash)**2
```

## pHash値を計算し比較する２

```python
lenna2 = PIL.Image.open('./data/DWT_lenna2.jpg')
```

```python
lenna2
```

* オリジナルから矩形切り抜きをし、テキストを追加した画像(lenna2)

```python
h2 = imagehash.phash(lenna2)
```

```python
print(h-h2)
```

```python
print(h2.hash)
```

numpyで[[8bit]x8]の形のリストとなっている

```python
print(len(h.hash))
```

リストの長さ返してしまうので下記で’**2’している

```python
print((h - h2)/len(h.hash)**2)
```

pHashのビット長に対する、ビット差の割合

### ハッシュ値の確認

```python
print(h)
print(h2)
```

```python
h_0x =  '0x'+str(h)
h2_0x =  '0x'+str(h2)
h_0b = format(int(h_0x, 0), '064b')
h2_0b = format(int(h2_0x, 0), '064b')
print(h_0b)
print(h2_0b)
```

```python
print(h-h2) #これはビット演算の差’^’の結果である
```

# aHashでのハッシュ値の差

```python
a = imagehash.average_hash(lenna)
a1 = imagehash.average_hash(lenna1)
a2 = imagehash.average_hash(lenna2)
```

```python
print((a-a1)/len(a.hash)**2)
```

```python
print((a-a2)/len(a.hash)**2)
```

## ハッシュ値の確認

```python
print(a)
print(a2)
```

```python
a_0x =  '0x'+str(a)
a2_0x =  '0x'+str(a2)
a_0b = format(int(a_0x, 0), '064b')
a2_0b = format(int(a2_0x, 0), '064b')
print(a_0b)
print(a2_0b)
```

```python
print(a-a1) #lenna vs. lenna1　のaHash値の差
print(a-a2) #lenna vs. lenna2
```

# wHash(DWT)でのハッシュ値の差

```python
w = imagehash.whash(lenna)#ここ各々でライブラリからwHash値が返されている
w1 = imagehash.whash(lenna1)
w2 = imagehash.whash(lenna2)
```

```python
print((w-w1)/len(w.hash)**2)#(ハミング距離/64bit)の形になっている。
```

```python
print((w-w2)/len(w.hash)**2)
```

## ハッシュ値の確認

```python
print(w)
print(w2)
```

```python
w_0x =  '0x'+str(w)
w2_0x =  '0x'+str(w2)
w_0b = format(int(w_0x, 0), '064b')
w2_0b = format(int(w2_0x, 0), '064b')
print(w_0b)
print(w2_0b)
```

```python
print(w-w1) #lenna vs. lenna1　のwHash値のハミング距離
print(w-w2) #lenna vs. lenna2　のwHash値のハミング距離
```

# コードの実行結果とWEBページの結果の比較

ここまでの各ハッシュ値計算の結果を表にして比較する

```python
import pandas as pd
import numpy as np
```

```python
compare = np.array([[(a-a1)/len(a.hash)**2, (a-a2)/len(a.hash)**2, (a-a_b)/len(a.hash)**2],
                                   [(h - h1)/len(h.hash)**2, (h - h2)/len(h.hash)**2, (h-h_b)/len(h.hash)**2],
                                   [(w-w1)/len(w.hash)**2, (w-w2)/len(w.hash)**2, (w-w_b)/len(w.hash)**2]])
```

```python
compare*100
```

```python
df = pd.DataFrame(compare*100, columns=['lenna vs. lenna1', 'lenna vs. lenna2', 'lenna vs. barbara'], index=['aHash', 'pHash', 'wHash'])
```

```python
df.T
```

小数第一まで丸め込んだ数字と一致している。
