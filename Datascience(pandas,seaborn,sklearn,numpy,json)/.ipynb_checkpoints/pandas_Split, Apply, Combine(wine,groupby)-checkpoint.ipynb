```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe_wine = pd.read_csv('winequality-red.csv', sep=';')
```

```python
dframe_wine.head()
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort_values('alcohol', ascending=False, inplace=True)
```

```python
quality_group = dframe_wine.groupby('quality')
quality_group.groups
```

```python
quality_group.get_group(3)
```

```python
#np.arange(len(quality_group.get_group(5))) + 1
```

```python
dframe_wine = quality_group.apply(ranker)
```
