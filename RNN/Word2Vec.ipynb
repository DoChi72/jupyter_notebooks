```python
from janome.tokenizer import Tokenizer
from gensim.models import word2vec
```

```python
import re
import zipfile
```

```python
localfile = "462_ruby_716.zip"
zipfile = zipfile.ZipFile(localfile, 'r')
file = zipfile.open('kazeno_matasaburo.txt', 'r')
bindata = file.read()
textdata = bindata.decode('shift_jis')
```

```python
textdata
```

```python
textdata = re.split(r'\-{5,}', textdata)[2]
```

```python
textdata = re.split(r'底本：', textdata)[0]
```

```python
textdata = textdata.strip()
```

```python
#形態素解析を行う
t = Tokenizer()
results = []
```

```python
lines = textdata.split("\r\n")
lines
```

```python
for line in lines:
    s = line
    s = s.replace('|','')
    s = re.sub(r'《.+?》','',s)
    s = re.sub(r'［＃.+?］','',s)
    tokens = t.tokenize(s)
    r = []
    for token in tokens:
        if token.base_form == "*":
            w = token.surface
        else:
            w = token.base_form
        ps = token.part_of_speech
        hinshi = ps.split(',')[0]
        if hinshi in ['名詞','形容詞','動詞','記号']:
            r.append(w)
    rl = (" ".join(r)).strip()
    results.append(rl)
    print(rl)
```

```python
#分かち書きテキストデータを保存
wakachigaki_file = "matasaburo.wakachi"
with open(wakachigaki_file, 'w', encoding='utf-8') as fp:
    fp.write("\n".join(results))
```

```python
#word2vec
#スキップグラムを利用した場合
data = word2vec.LineSentence(wakachigaki_file)
model = word2vec.Word2Vec(data, size=200, window=5, hs=1, min_count=2, sg=1)
model.save('matasaburo.model')
```
