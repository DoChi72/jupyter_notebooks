```python
import numpy as np
import tensorflow as tf
```

```python
with open('./reviews.txt', 'r') as f:
    reviews = f.read()
```

```python
reviews[0:100]
```

```python
with open('./labels.txt', 'r') as f:
    labels = f.read()
```

```python
labels[0:100]
```

```python
from string import punctuation
all_text = ''.join([c for c in reviews if c not in punctuation])
reviews = all_text.split('\n')

all_text = ' '.join(reviews)
words = all_text.split()
```

```python
#all_text[:200]
```

```python
#reviews[:200]
```

```python
print(sorted(set(words[:200])))
```

```python
#単語を切り出し
from collections import Counter
counts = Counter(words)
vocab = sorted(counts, key=counts.get, reverse=True) 
```

```python
#単語と番号の辞書
vocab_to_int = {word:ii for ii, word in enumerate(vocab, 1)}
```

```python
reviews_int = []
for each in reviews:
    reviews_int.append([vocab_to_int[word] for word in each.split()])
```

```python
reviews_int[:10]
```

```python
#ラベルをベクトルにする
labels = labels.split('\n')
labels = np.array([1 if each =='positive' else 0 for each in labels])
```

```python
review_lens = Counter([len(x) for x in reviews_int])
```

```python
review_lens[0]
```

```python
max(review_lens)
```

```python
non_zero_index = [ii for ii, review in enumerate(reviews_int) if len(review) != 0]
```

```python
len(non_zero_index)
```

```python
reviews_int[-1]
```

```python
reviews_int = [reviews_int[ii] for ii in non_zero_index]
```

```python
labels = np.array([labels[ii] for ii in non_zero_index])
```

```python
seq_len = 200
features = np.zeros((len(reviews_int), seq_len), dtype=int)
```

```python
for i, row in enumerate(reviews_int):
    features[i, -len(row):] = np.array(row)[:seq_len]
```

```python
features[:10,:100]
```

```python
#データを教師用とテスト用に分割する
split_frac = 0.8
split_index = int(len(features)*0.8)

train_x, val_x = features[:split_index], features[split_index:]
train_y, val_y = labels[:split_index], labels[split_index:]
```

```python
test_index = int(len(val_x)*0.5)
val_x, test_x = val_x[:test_index], val_x[test_index:]
val_y, test_y = val_y[:test_index], val_y[test_index:]
```

```python
print("Train set: \t\t{}".format(train_x.shape))
```

```python
print("Validation set: \t{}".format(val_x.shape))
```

```python
print("Test set: \t\t{}".format(test_x.shape))
```

```python
#エンベディングレイヤーの定義
lstm_size = 256
lstm_layers = 1
batch_size = 500
learning_rate = 0.001
```

```python
n_words = len(vocab_to_int) + 1

graph = tf.Graph()
with graph.as_default():
    inputs_ = tf.placeholder(tf.int32, [None, None], name='input')
    labels_ = tf.placeholder(tf.int32, [None, None], name='labels')
    keep_prob = tf.placeholder(tf.float32, name = 'keep_prob')
```

```python
embed_size = 300

with graph.as_default():
    embedding = tf.Variable(tf.random_uniform((n_words, embed_size), -1, 1))
    embed = tf.nn.embedding_lookup(embedding, inputs_)
```

```python
#LSTMセルとレイヤーの定義
with graph.as_default():
    lstm = tf.contrib.rnn.BasicLSTMCell(lstm_size)
    drop = tf.contrib.rnn.DropoutWrapper(lstm,output_keep_prob = keep_prob)
    cell = tf.nn.rnn_cell.MultiRNNCell([drop] * lstm_layers)
    initial_state = cell.zero_state(batch_size, tf.float32)
```

```python
with graph.as_default():
    outputs, final_state = tf.nn.dynamic_rnn(cell, embed, initial_state=initial_state)
```

```python
#推定値の計算と損失関数、最適化処理の定義
with graph.as_default():
    predictions = tf.contrib.layers.fully_connected(outputs[:,-1], 1, activation_fn=tf.sigmoid)
    cost = tf.losses.mean_squared_error(labels_, predictions)
    optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
#学習精度の計測
with graph.as_default():
    correct_pred = tf.equal(tf.cast(tf.round(predictions), tf.int32), labels_)
    accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
#バッチ処理
def get_batches(x, y, batch_size=100):
    n_batches = len(x)//batch_size
    x, y, = x[:n_batches*batch_size], y[:n_batches*batch_size]
    for ii in range(0, len(x), batch_size):
        yield x[ii:ii+batch_size], y[ii:ii+batch_size]
```

```python
#トレーニング
epochs = 5
with graph.as_default():
    saver = tf.train.Saver()
    
with tf.Session(graph=graph) as sess: #今までのグラフ定義を渡す
    sess.run(tf.global_variables_initializer())
    iteration = 1
    for e in range(epochs):
        state = sess.run(initial_state)
        for ii, (x, y) in enumerate(get_batches(train_x,train_y, batch_size),1):
            feed = {inputs_: x,
                   labels_: y[:,None],
                   keep_prob: 0.5,
                   initial_state: state}
            loss, state, _ = sess.run([cost, final_state, optimizer], feed_dict=feed)
            
            if iteration%5==0:
                print("Epoch: {}/{}".format(e, epochs),
                     "Iteration: {}".format(iteration),
                     "Training Loss: {:.3f}".format(loss))
            
            if iteration%25==0:
                val_acc = []
                val_state = sess.run(cell.zero_state(batch_size, tf.float32))
                for x, y in get_batches(val_x, val_y, batch_size):
                    feed = {inputs_: x,
                           labels_: y[:,None],
                           keep_prob: 1,
                           initial_state: val_state}
                    batch_acc, val_state = sess.run([accuracy, final_state], feed_dict=feed)
                    val_acc.append(batch_acc)
                print("Value Acc: {:.3f}".format(np.mean(val_acc)))
            iteration += 1
    saver.save(sess, "checkpoint/sentiment.ckpt")
```

```python
test_acc = []
with tf.Session(graph=graph) as sess:
    saver.restore(sess, tf.train.latest_checkpoint('checkpoint'))
    test_state = sess.run(cell.zero_state(batch_size, tf.float32))
    for ii , (x,y) in enumerate (get_batches(test_x,test_y, batch_size),1):
        feed = {inputs_: x,
                labels_: y[:,None],
                keep_prob: 1,
                initial_state: test_state}
        batch_acc, val_state = sess.run([accuracy, final_state], feed_dict=feed)
        test_acc.append(batch_acc)
    print("Test Accuracy: {:.3f}".format(np.mean(test_acc)))
```
